```python
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming you have already defined X_train, Y_train, X_test, and Y_test

# Step 5: Define the hyperparameter grid
param_grid = {
    'n_neighbors': np.arange(3, 21, 2),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'minkowski', 'manhattan']
}

# Step 6: Set up the GridSearchCV with progress monitoring
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

grid_search = GridSearchCV(
    knn,
    param_grid,
    cv=5,
    scoring='accuracy',
    return_train_score=True,
    verbose=3,  # Verbose level 3 for detailed progress updates
    n_jobs=-1   # Use all available processors
)

# Step 7: Fit the GridSearchCV on the training data
print("Starting GridSearchCV...")
grid_search.fit(X_train, Y_train)

# Step 8: Extract the results into a pandas DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# Step 9: Convert categorical columns to numerical for plotting
results['weights'] = results['param_weights'].map({'uniform': 0, 'distance': 1})
results['metric'] = results['param_metric'].map({'euclidean': 0, 'manhattan': 1, 'minkowski': 2})

# Step 10: Set up 3D plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Step 11: Extract hyperparameters and score for plotting
x = results['param_n_neighbors'].astype(int)
y = results['weights'].astype(int)  # 0 = uniform, 1 = distance
z = results['metric'].astype(int)  # 0 = euclidean, 1 = manhattan, 2 = minkowski
accuracy = results['mean_test_score']

# Step 12: Plot the 3D scatter plot
img = ax.scatter(x, y, z, c=accuracy, cmap='viridis', s=100)

# Step 13: Customize the plot
ax.set_xlabel('Number of Neighbors (k)')
ax.set_ylabel('Weights (0=uniform, 1=distance)')
ax.set_zlabel('Metric (0=euclidean, 1=manhattan, 2=minkowski)')
ax.set_title('3D Plot of Hyperparameters and Cross-Validation Accuracy')

# Add color bar
cbar = fig.colorbar(img, ax=ax, pad=0.1)
cbar.set_label('Cross-Validation Accuracy')

plt.savefig('knn_hyperparameter_tuning.pdf', dpi=300)

plt.show()

# Print the best parameters and their corresponding accuracy
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.2f}")

# Step 14: Evaluate on the test set with the best parameters
best_knn = grid_search.best_estimator_
test_accuracy = best_knn.score(X_test, Y_test)
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Define the SVM model
svm_model = SVC()  # No need to specify kernel here, as GridSearchCV will test multiple

# Define the hyperparameter grid
param_grid = {
    'C': [0.1, 1, 10],  # Regularization parameter
    'kernel': ['linear', 'rbf', 'sigmoid'],  # SVM kernel types
    'gamma': ['scale', 'auto', 0.1]  # Kernel coefficient
    'degree': [2, 3, 4]  # Only used for 'poly' kernel
}

# Perform Grid Search with Cross Validation
grid_search = GridSearchCV(
    estimator=svm_model, 
    param_grid=param_grid, 
    scoring='accuracy',  # Ensure this is a valid metric
    cv=5, 
    verbose=2, 
    n_jobs=-1
)

# Fit the model to the data
grid_search.fit(X_train, Y_train)

# Get the best parameters and the best score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Accuracy: {grid_search.best_score_:.2f}")

# Evaluate the model on the test set
best_model = grid_search.best_estimator_
y_pred_svm = best_model.predict(X_test)

# Print Classification Report
print("\nClassification Report:\n", classification_report(Y_test, y_pred_svm))
```

```python
# Initialize the model
model = Sequential()
model.add(Dense(units=128, activation='relu', input_shape=(X_train.shape[1],)))

# Add hidden layers
model.add(Dense(units=128, activation='relu'))
# model.add(Dense(units=128, activation='relu'))

model.add(Dropout(0.3))

# Add output layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the model
history = model.fit(X_train, Y_train, epochs=50, batch_size=256, validation_split=0.2)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, Y_test)


print(f'Accuracy: {test_accuracy}, Loss: {test_loss}')
```
